# TASK 1 - PREDICTION USING SUPERVISED ML

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('http://bit.ly/w-data')
data.head()
```

```python
data.isnull == True
```

```python
plt.title("Score Vs Hours", fontsize = 10)
plt.xlabel("Hours", fontsize = 15)
plt.ylabel("Score(in %)", fontsize = 15)
sns.scatterplot(y = data['Scores'], x = data['Hours'] )
sns.set_style('darkgrid')
```

```python
plt.title("Regression Plot", fontsize = 10)
plt.xlabel("Hours", fontsize = 15)
plt.ylabel("Score(in %)", fontsize = 15)
sns.regplot(y = data['Scores'], x = data['Hours'])
plt.show()
```

```python
print(data.corr())
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,1].values
train_X, test_X, train_y, test_y = train_test_split(X,y, random_state = 0 ) 
```

```python
regression = LinearRegression()
regression.fit(train_X, train_y)
print("Training Completed Successfully!")
```

```python
pred_y = regression.predict(test_X)
print(pred_y)
```

```python
test_y
```

```python
test_A = list(test_y)
prediction = list(pred_y)
df_compare = pd.DataFrame({'Hours':test_A, 'Predicted Score':prediction})
df_compare
```

```python
plt.scatter(x = test_X, y = test_y, color = 'Red')
plt.plot(test_X, pred_y, color = 'Black')
plt.xlabel("Hours", fontsize = 15)
plt.ylabel("Score( in %)", fontsize = 15)
plt.title("Actual Score Vs Predicted Score", fontsize = 20)
```

```python
print("Absolute Mean Error:", mean_absolute_error(test_X, pred_y))
```

```python
hours = [9.25]
solution = regression.predict([hours])
print("Score{}".format(round(solution[0],3)))
```
